```python
samples = {'SCRN_PPMI3186_2662':[0.268940683,0.270072616],
           'SCRN_PPMI3220_6139':[0.218830678,0.21895899,0.219054752,0.218684714],
           'SCRN_PPMI18567_4073':[0.183450781,0.184177792],
           'SCRN_PPMI3234_1338':[0.216650102,0.21777278],
           'SCRN_PPMI3409_0660':[0.219620145,0.220405266],
           'SCRN_PPMI3419_0298':[0.20247863,0.203395931,0.202775047,0.202310672],
           'SCRN_PPMI3422_1260':[0.21255287,0.212883121,0.21315043],
           'SCRN_PPMI3411_0083':[0.261062085,0.261657395,0.262684855,0.26161835],
           'SCRN_PPMI3452_7426':[0.22096759,0.222082429],
           'SCRN_PPMI3453_7504':[0.226412118,0.227765643],
           'SCRN_PPMI3460_1783':[0.250243633,0.251433532,0.250653278,0.250635374]
          }
```

```python
mediancolor="lightseagreen"
```

```python
samp_avgs = {}

def avg(lsamp):
    cnt = 0
    tot = 0
    for val in lsamp:
        cnt += 1
        tot += val
    return tot/cnt

for samp in samples:
    samp_avgs[samp] = [avg(samples[samp])]
```

```python
agarwal = [0.3942895325790199, 0.45196678897537274, 0.42483972982519896, 0.3977922434126851, 0.41728344189596756, 0.39828376914541724, 0.4160610371455113]
wc = []
for samp in samples:
    wc = wc + samp_avgs[samp]

labs = ['whole cell', 'single nucleus']
testys = [wc, agarwal]

plt.figure(figsize=(3,6))
plt.boxplot(testys, tick_labels=labs, medianprops={'color': mediancolor}, widths=0.5)
plt.xticks(rotation=90)
plt.ylim([0,.5])
plt.ylabel('Fraction of introns retained')
plt.rcParams.update({'font.size': 16})

plt.show()
```
