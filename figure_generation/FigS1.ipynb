```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f1s = {}
wholecell = "#990000"
nucseq = "#038a80"
```

```python
f1s['genes']['0.2']=[0.84864,0.86837,0.86805,0.86805]
f1s['genes']['0.4']=[0.84822,0.86517,0.86491,0.86491]
```

```python
xs = [1,10,100,1000]
```

```python
plt.plot(xs,f1s['genes']['0.2'], '.-', color=wholecell, label='whole cell', alpha=0.7)
plt.plot(xs,f1s['genes']['0.4'], '.-', color=nucseq,label='single nucleus', alpha=0.7)
plt.legend(loc='lower right', prop={'size':12})
plt.rcParams.update({'font.size': 16})
plt.xscale('log')
plt.ylim([0.5,1.0])
plt.title('Accuracy on genes')
plt.xlabel('Mapping locations allowed')
plt.ylabel('F1 score')
plt.savefig('25.11.07_supp_AA.png', bbox_inches="tight", pad_inches=0.2)
plt.show()
```
